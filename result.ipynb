```python
import os
from IPython.display import Audio,display 
import numpy as np
# audio
path = './result/'
i=1
for parent,dirnames,_ in os.walk(path):
    for dir in dirnames:
        print '============== %d ================='%i
        for file_name in os.listdir(path+dir):
            if file_name.endswith(".mp3"):
                if 'origin' in file_name.strip().split('.'):
                    print 'Origin melody'
                else:
                    print 'Generated result'
                display(Audio(path+dir+'/'+file_name)) 
        i+=1

```
